```python
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

# Initialize Faker for generating synthetic names and emails
fake = Faker()

# Number of records to generate for each table
NUM_CUSTOMERS = 10
NUM_BRANCHES = 3
NUM_ORDERS = 20
NUM_MENU_ITEMS = 5
NUM_PAYMENT_METHODS = 4

# Create data for Customers table
customers_data = {
    "Customer_ID": range(1, NUM_CUSTOMERS + 1),
    "Customer_Name": [fake.name() for _ in range(NUM_CUSTOMERS)],
    "Phone_Number": [fake.phone_number() for _ in range(NUM_CUSTOMERS)],
    "Email": [fake.email() for _ in range(NUM_CUSTOMERS)],
    "Customer_Type": [random.choice(["Regular", "Premium", "VIP"]) for _ in range(NUM_CUSTOMERS)]
}
customers_df = pd.DataFrame(customers_data)

# Create data for Branch table
branch_data = {
    "Branch_ID": range(1, NUM_BRANCHES + 1),
    "Branch_Name": [f"{fake.city()} Branch" for _ in range(NUM_BRANCHES)],
    "Branch_Address": [fake.address() for _ in range(NUM_BRANCHES)],
    "City": [fake.city() for _ in range(NUM_BRANCHES)],
    "State": [fake.state() for _ in range(NUM_BRANCHES)],
    "Zip_Code": [fake.zipcode() for _ in range(NUM_BRANCHES)]
}
branch_df = pd.DataFrame(branch_data)

# Create data for Payment_Method table
payment_method_data = {
    "Payment_Method_ID": range(1, NUM_PAYMENT_METHODS + 1),
    "Payment_Type": ["Credit Card", "Debit Card", "PayPal", "Cash"]
}
payment_method_df = pd.DataFrame(payment_method_data)

# Create data for Menu_Items table
menu_items_data = {
    "Menu_Item_ID": range(1, NUM_MENU_ITEMS + 1),
    "Menu_Item_Name": ["Fufu with Egusi Soup", "Jollof Rice with Chicken", "Pounded Yam with Vegetable Soup", "Suya", "Plantain Chips"],
    "Category": ["Main", "Main", "Main", "Starter", "Starter"],
    "Price": [15.00, 12.50, 14.00, 7.00, 5.00]
}
menu_items_df = pd.DataFrame(menu_items_data)

# Create data for Inventory table
inventory_data = {
    "Inventory_ID": range(1, NUM_MENU_ITEMS * NUM_BRANCHES + 1),
    "Branch_ID": [random.randint(1, NUM_BRANCHES) for _ in range(NUM_MENU_ITEMS * NUM_BRANCHES)],
    "Menu_Item_ID": [random.randint(1, NUM_MENU_ITEMS) for _ in range(NUM_MENU_ITEMS * NUM_BRANCHES)],
    "Stock_Level": [random.randint(50, 200) for _ in range(NUM_MENU_ITEMS * NUM_BRANCHES)]
}
inventory_df = pd.DataFrame(inventory_data)

# Create data for Orders table
orders_data = {
    "Order_ID": range(1, NUM_ORDERS + 1),
    "Order_Date": [(datetime.now() - timedelta(days=random.randint(0, 30))).strftime("%Y-%m-%d") for _ in range(NUM_ORDERS)],
    "Order_Time": [(datetime.now() - timedelta(hours=random.randint(1, 24))).strftime("%H:%M:%S") for _ in range(NUM_ORDERS)],
    "Branch_ID": [random.randint(1, NUM_BRANCHES) for _ in range(NUM_ORDERS)],
    "Customer_ID": [random.randint(1, NUM_CUSTOMERS) for _ in range(NUM_ORDERS)],
    "Payment_Method_ID": [random.randint(1, NUM_PAYMENT_METHODS) for _ in range(NUM_ORDERS)],
    "Dining_Option": [random.choice(["Dine-In", "Delivery", "Takeaway"]) for _ in range(NUM_ORDERS)]
}
orders_df = pd.DataFrame(orders_data)

# Create data for Order_Items table
order_items_data = {
    "Order_Items_ID": range(1, NUM_ORDERS + 1),
    "Order_ID": [random.randint(1, NUM_ORDERS) for _ in range(NUM_ORDERS)],
    "Menu_Item_ID": [random.randint(1, NUM_MENU_ITEMS) for _ in range(NUM_ORDERS)],
    "Quantity": [random.randint(1, 5) for _ in range(NUM_ORDERS)],
    "Unit_Price": [random.choice([15.00, 12.50, 14.00, 7.00, 5.00]) for _ in range(NUM_ORDERS)]
}
order_items_df = pd.DataFrame(order_items_data)

# Save all tables to CSV files
customers_df.to_csv('customers.csv', index=False)
branch_df.to_csv('branch.csv', index=False)
payment_method_df.to_csv('payment_method.csv', index=False)
menu_items_df.to_csv('menu_items.csv', index=False)
inventory_df.to_csv('inventory.csv', index=False)
orders_df.to_csv('orders.csv', index=False)
order_items_df.to_csv('order_items.csv', index=False)

print("CSV files generated successfully!")
```

```python
import os
import pandas as pd
from sqlalchemy import create_engine

# PostgreSQL connection details
USER = 'postgres'         # Replace with your PostgreSQL username
PASSWORD = 'blossom'     # Replace with your PostgreSQL password
HOST = 'localhost'             # Replace with your host, typically localhost
PORT = '5432'                  # Replace with your PostgreSQL port, default is 5432
DATABASE = 'fufu_republic'     # Replace with your database name

# Create the connection string
connection_string = f'postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}'

# Create the SQLAlchemy engine
engine = create_engine(connection_string)

# Folder containing the CSV files
folder_path = "C:\\Users\\Qucoon Limited\\Desktop\\CDE\\dbt"  # Replace with the path to your folder

# Function to clean and create table names from CSV filenames
def clean_table_name(file_name):
    table_name = os.path.splitext(file_name)[0]  # Remove the .csv extension
    table_name = table_name.lower().replace(' ', '_')  # Lowercase and replace spaces with underscores
    return table_name

# Iterate through each file in the folder
for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        table_name = clean_table_name(file)
        print(f'Processing file: {file} into table: {table_name}')

        # Load the CSV into a DataFrame
        df = pd.read_csv(file_path)

        # Load the DataFrame into PostgreSQL
        try:
            # Write the DataFrame to PostgreSQL, replacing or appending as needed
            df.to_sql(table_name, engine, index=False, if_exists='replace')  # Use 'append' to add data without dropping the table
            print(f'Successfully loaded {file} into {table_name}')
        except Exception as e:
            print(f'Error loading {file} into {table_name}: {e}')

# Close the connection
engine.dispose()
```

```python
import psycopg2
from faker import Faker
import random
from datetime import datetime, timedelta

# Initialize Faker for generating synthetic data
fake = Faker()

# PostgreSQL connection details
db_config = {
    "dbname": "postgres",
    "user": "postgres",
    "password": "C0met0me123$",
    "host": "localhost",
    "port": "5432"
}

# Connect to the PostgreSQL database
conn = psycopg2.connect(**db_config)
cur = conn.cursor()

# Number of records to generate
NUM_CUSTOMERS = 10
NUM_BRANCHES = 3
NUM_ORDERS = 20
NUM_MENU_ITEMS = 5
NUM_PAYMENT_METHODS = 4

# Insert Customers data
for _ in range(NUM_CUSTOMERS):
    cur.execute("""
        INSERT INTO Customers (Customer_Name, Phone_Number, Email, Customer_Type)
        VALUES (%s, %s, %s, %s)
    """, (fake.name(), fake.phone_number(), fake.email(), random.choice(["Regular", "Premium", "VIP"])))

# Insert Branch data
for _ in range(NUM_BRANCHES):
    cur.execute("""
        INSERT INTO Branch (Branch_Name, Branch_Address, City, State, Zip_Code)
        VALUES (%s, %s, %s, %s, %s)
    """, (f"{fake.city()} Branch", fake.address(), fake.city(), fake.state(), fake.zipcode()))

# Insert Payment_Method data
payment_methods = ["Credit Card", "Debit Card", "PayPal", "Cash"]
for method in payment_methods:
    cur.execute("""
        INSERT INTO Payment_Method (Payment_Type)
        VALUES (%s)
    """, (method,))

# Insert Menu_Items data
menu_items = [
    ("Fufu with Egusi Soup", "Main", 15.00),
    ("Jollof Rice with Chicken", "Main", 12.50),
    ("Pounded Yam with Vegetable Soup", "Main", 14.00),
    ("Suya", "Starter", 7.00),
    ("Plantain Chips", "Starter", 5.00)
]
for item in menu_items:
    cur.execute("""
        INSERT INTO Menu_Items (Menu_Item_Name, Category, Price)
        VALUES (%s, %s, %s)
    """, item)

# Insert Inventory data
for branch_id in range(1, NUM_BRANCHES + 1):
    for menu_item_id in range(1, NUM_MENU_ITEMS + 1):
        cur.execute("""
            INSERT INTO Inventory (Branch_ID, Menu_Item_ID, Stock_Level)
            VALUES (%s, %s, %s)
        """, (branch_id, menu_item_id, random.randint(50, 200)))

# Insert Orders data
for _ in range(NUM_ORDERS):
    order_date = (datetime.now() - timedelta(days=random.randint(0, 30))).strftime("%Y-%m-%d")
    order_time = (datetime.now() - timedelta(hours=random.randint(1, 24))).strftime("%H:%M:%S")
    cur.execute("""
        INSERT INTO Orders (Order_Date, Order_Time, Branch_ID, Customer_ID, Payment_Method_ID, Dining_Option)
        VALUES (%s, %s, %s, %s, %s, %s)
    """, (
        order_date,
        order_time,
        random.randint(1, NUM_BRANCHES),
        random.randint(1, NUM_CUSTOMERS),
        random.randint(1, NUM_PAYMENT_METHODS),
        random.choice(["Dine-In", "Delivery", "Takeaway"])
    ))

# Insert Order_Items data
cur.execute("SELECT Order_ID FROM Orders")
order_ids = cur.fetchall()

for order_id in order_ids:
    for _ in range(random.randint(1, 3)):  # Random number of items per order
        menu_item_id = random.randint(1, NUM_MENU_ITEMS)
        quantity = random.randint(1, 5)
        cur.execute("""
            INSERT INTO Order_Items (Order_ID, Menu_Item_ID, Quantity, Unit_Price)
            VALUES (%s, %s, %s, %s)
        """, (order_id, menu_item_id, quantity, random.choice([15.00, 12.50, 14.00, 7.00, 5.00])))

# Commit the changes and close the connection
conn.commit()
cur.close()
conn.close()

print("Data populated successfully!")
```
